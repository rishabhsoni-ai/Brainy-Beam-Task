TASK 5 :- Generate a parallel coordinates plot for visualizing high-dimensional data.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from pandas.plotting import parallel_coordinates

np.random.seed(42)
df = pd.DataFrame({
    "Feature1": np.random.randint(10, 50, 10),
    "Feature2": np.random.randint(100, 200, 10),
    "Feature3": np.random.rand(10) * 10,
    "Feature4": np.random.randn(10) * 5 + 50,
    "Class": ["A","A","A","B","B","B","C","C","C","C"]
})

print("Sample DataFrame:")
print(df)

plt.figure(figsize=(10,6))
parallel_coordinates(df, "Class", colormap=plt.cm.Set1)
plt.title("Parallel Coordinates Plot (Matplotlib)")
plt.grid(True)
plt.show()

import plotly.express as px

df['Class_numerical'] = df['Class'].astype('category').cat.codes

fig = px.parallel_coordinates(
    df,
    color="Class_numerical",
    dimensions=["Feature1", "Feature2", "Feature3", "Feature4"],
    color_continuous_scale=px.colors.diverging.Tealrose,
    labels={"Class_numerical": "Class"}
)
fig.update_layout(title="Interactive Parallel Coordinates Plot (Plotly)")
fig.show()
```
