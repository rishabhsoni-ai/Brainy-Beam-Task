TASK 2 :- Implement a custom colormap for a Matplotlib plot.

```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.colors import LinearSegmentedColormap, ListedColormap, TwoSlopeNorm


x = np.linspace(-3, 3, 100)
y = np.linspace(-3, 3, 100)
X, Y = np.meshgrid(x, y)
Z = np.sin(X**2 + Y**2)


colors = ["darkblue", "cyan", "yellow", "red"]
custom_cmap = LinearSegmentedColormap.from_list("mycmap", colors, N=256)

plt.figure(figsize=(6,5))
plt.imshow(Z, cmap=custom_cmap, origin="lower", extent=[-3,3,-3,3])
plt.colorbar(label="Value")
plt.title("Custom Colormap (LinearSegmentedColormap)")
plt.show()


discrete_cmap = ListedColormap(["#2E8B57", "#FFD700", "#FF4500"])

plt.figure(figsize=(6,5))
plt.contourf(X, Y, Z, levels=3, cmap=discrete_cmap)
plt.colorbar(label="Category")
plt.title("Custom Colormap (ListedColormap - Discrete)")
plt.show()


norm = TwoSlopeNorm(vmin=-1, vcenter=0, vmax=1)

plt.figure(figsize=(6,5))
plt.imshow(Z, cmap=custom_cmap, norm=norm, origin="lower", extent=[-3,3,-3,3])
plt.colorbar(label="Centered at 0")
plt.title("Custom Colormap with Normalization")
plt.show()
```
